### Importing an Excel with Pandas <a class="anchor" id="Importing-an-Excel-with-Pandas"></a>

```python
# setting working directory
os.chdir(r'C:\Users\Anant Awasthi\OneDrive - Amity University\Value Added Certification Course')
```

```python
# Importing Data from different sheets
School_data = pd.read_excel('Data/School Data.xlsx')
School_data['Pincode'] = School_data.Pincode.apply(lambda x: str(x))
```

### Charts
* Bar Diagram
* Pie Chart
* Box Plot
* Line Chart

### Bar Chart

```python
School_summary = School_data.groupby('Pincode').aggregate(SchoolCount=('Schools', sum)).reset_index()
School_summary
```

### Building a raw visualization

```python

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(School_summary.Pincode,School_summary.SchoolCount)
plt.show()
```

### Adding a Title to the Chart

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(School_summary.Pincode,School_summary.SchoolCount)
plt.title('Analysis of School by Pincode')
plt.show()
```

### Adding axis Labels

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(School_summary.Pincode,School_summary.SchoolCount)
plt.title('Analysis of School by Pincode')
plt.ylabel('Count of Schools')
plt.xlabel('Pincode')
plt.show()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
plt.pie(School_summary.SchoolCount, labels = School_summary.Pincode)
plt.title('Distribution of School by Pincode')
plt.show() 
```
