#### What are functions or User Defined Functions (UDF)?
In Python, functions are named blocks of reusable code that perform a specific task. They allow you to break down a complex program into smaller, manageable pieces, improving code organization, reusability, and readability. Functions take input arguments, perform operations based on the provided arguments, and can return output values.

Here are some key characteristics and benefits of functions in Python:

* **Code Reusability**: Functions can be defined once and called multiple times from different parts of a program, eliminating code duplication and promoting reusability.

* **Modularity**: Functions enable you to divide a program into smaller, self-contained modules, each responsible for a specific task. This makes the code more organized, easier to understand, and maintainable.

* **Encapsulation**: Functions encapsulate a set of operations, allowing you to hide the implementation details and work with a higher-level abstraction. This enhances code readability and reduces complexity.

### Using Dataframe as an orgument in a function

#### Definition of An UDF

```python
import pandas as pd

def double_column_values(df, column_name):
    df[column_name] = df[column_name] * 2
    return df
```

#### Creating a Dataframe

```python
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
df
```

#### Calling function on a pandas Dataframe

```python
modified_df = double_column_values(df, 'A')
print(modified_df)
```

### Calling function in a Dataframe 

#### Definition of An UDF

```python
import pandas as pd

def square_value(x):
    return x ** 2
```

#### Creating a Dataframe

```python
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
df
```

#### Calling function on a pandas Dataframe field

```python
df['Function_Output'] = df['A'].apply(square_value)
df
```
