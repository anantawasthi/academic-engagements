#### What are lambda functions?
Lambda functions, also known as anonymous functions, are one-line functions in Python that are defined without a name using the lambda keyword. They are useful when you need a simple function for a specific operation and don't want to define a full-fledged function using def. Lambda functions are often used in combination with built-in functions like map(), filter(), and reduce(), or as short, inline functions.

#### Lambda functions in Python

Here's a quick overview of writing lambda Functions in Python along with example:
* Defining lambda Functions
* lambda Functions Parameters
* lambda Functions Code Block


<p></p>

#### Definition of Lambda Function

```python
square = lambda x: x ** 2
```

#### Calling a function

```python
result = square(5)
result  # Output: 25
```

In this example, the lambda function lambda x: x ** 2 takes an argument x and returns the square of x using the exponentiation operator **. The lambda function is assigned to the variable square, and it is then called with the argument 5 to calculate the square, which is assigned to the variable result.

lambda functions are a better choice when simplicity, conciseness, and inline usage are desired, particularly for small-scale operations or when working within the functional programming paradigm. They provide a convenient way to create short, throwaway functions without the need for a dedicated function definition.

### Using Lambda Functions in Pandas

#### Creating a Dataframe

```python
import pandas as pd
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
df
```

#### Calling function on a pandas Dataframe

```python
# Applying a lambda function to each element of a column
df['A_Square'] = df['A'].apply(lambda x: x ** 2)
df
```
